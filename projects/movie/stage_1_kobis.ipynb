```python
import os

key = os.getenv("KOBIS_KEY")
```

```python
import requests
import datetime
import csv

# 20190113으로 시간 설정
current = datetime.datetime.now()
base_week = current - datetime.timedelta(days = 5)

#일주일 간격으로 10번 반복문으로 정보수집
movie_names = []
data_dict = {}
for a in range(0,10):
    target_week = base_week - datetime.timedelta(weeks = (9-a))
    target_week = target_week.strftime("%Y%m%d")
#     print(target_week)
    url = f"http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchWeeklyBoxOfficeList.json?key={key}&targetDt={target_week}&weekGb=0"
    data = requests.get(url).json()
    data_set = data["boxOfficeResult"]["weeklyBoxOfficeList"]
    week_time = data["boxOfficeResult"]["yearWeekTime"]
# i는 순위 10개 까지 정보 보여줌
    for i in range(10):
        if data_set[i]["movieNm"] not in movie_names:
            movie_names.append(data_set[i]["movieNm"])
            data_dict[data_set[i]["movieNm"]] = [data_set[i]["movieCd"], data_set[i]["movieNm"], data_set[i]["audiAcc"] ,target_week]
        else:
            data_dict[data_set[i]["movieNm"]][2] = data_set[i]["audiAcc"]
            data_dict[data_set[i]["movieNm"]][3] = target_week     
```

```python
f_w = open("boxoffice.csv", "w+", encoding="utf-8", newline="")            

writer = csv.writer(f_w)

for value in data_dict.values():
    writer.writerow(value)

f_w.close()
```
