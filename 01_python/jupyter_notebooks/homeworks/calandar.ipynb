```python
last_day_check = 0

# 무슨 달인지 표시하는 부분

for month_number in range(1,13):
    print("**************************")
    print("           {} 월".format(month_number))
    print("--------------------------")
    print("일  월  화  수  목  금  토")
    print("--------------------------")
    
# 한 달 일수 구하는 부분

    if month_number == 2:
        number_of_days = 28
    elif month_number <= 7:
        number_of_days = 30 + month_number % 2
    else:
        number_of_days = 31 - month_number % 2
        
        
# 저번달 마지막 주 일수와 일주일 단위로 줄 바꾸게 해줄 변수를 연결하고, 반복문 안에 들어갈 변수 초기화하는 부분
    
    week_changer = last_day_check
    days_of_last_week1 = week_check = 0
    days_of_last_week2 = 1
    for t in range(week_changer):
        if week_changer == 7:
            break
        print(" ", end = "   ")

# 날짜를 출력하고, 저번달 일 수 + 날짜가 7의 배수일때마다 줄바꾸는 부분  

    for days in range(1,number_of_days+1):
        if days >= 10:
            print(days, end = "  ")
        else:
            print(days, end = "   ")
        week_changer += 1
        if week_changer % 7 == 0 and days == number_of_days:
            week_check += 1
        elif week_changer % 7 == 0:
            week_check += 1
            print("")
        if week_check == 4:
            days_of_last_week1 += 1
        if week_check == 5:
            days_of_last_week2 += 1
             
            
# 달 사이에 간격을 추가하는 부분  

    print("\n")
    print("**************************")
    
    
    
# 마지막 주 일수 구하는 부분
    if days_of_last_week2 >= 2:
        last_day_check = days_of_last_week2 - 2
    else:
        last_day_check = days_of_last_week1 - 1
```
