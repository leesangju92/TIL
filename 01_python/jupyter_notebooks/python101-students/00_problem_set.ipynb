## 1. 이상한 덧셈
> 숫자로 구성된 리스트에서 양의 정수의 합을 구하는 함수, `positive_sum`을 만들어 보아요

예시)
``` python
positive_sum([1, -10, 2]) #3
positive_sum([-1, -2, -3, -4]) #0
```

````python
def positive_sum(number_list):
    sum_of_number = 0
    for number in number_list:
        if number > 0:
            sum_of_number += number
    return sum_of_number

# def pos_sum(numbers):
#     return sum(x for x in numbers if x >0 )
            
positive_sum([-1, -2, -3, -4])
````

### 문자열 탐색
> 문자열 요소로만 이루어진 리스트를 넣었을 때, 문자열의 길이가 2이상이고 주어진 문자열의 첫번째와 마지막 문자가 같은 문자열의 수를 세는 함수 `start_end`를 작성하세요

예시)
```python
start_end(['level', 'asdwe', 's', 'abadsfa', 'qiq']) #n=3
```

````python
def start_end(word_list):
    n = 0
    for word in word_list:
        if list(word)[0] == list(word)[-1] and len(word) >= 2:
            n += 1
    return n  


start_end(['level', 'asdwe', 'sss', 'abadsfa', 'qiq'])
````

### Coliatz
> coliatz 추측 : 어떤 자연수 n 이던지, 다음과 같은 작업을 반복하면 i로 만들수 있다.

1. n이 짝수라면 2로 나눈다
2. n이 홀수라면 3을 곱하고 1을 더한다
3. 결과로 나온 수에 1 / 2의 작업을 1이 될때까지 반복한다

>예를 들어 n이 6이면, 6 ==> 3 ==> 10 ==> 5 ==> 16 ==> 8 ==> 4 ==> 2 ==> 1이며 8번(>=갯수!)만에 1이 됩니다. 자연수 n이 들어왔을 때, 몇 번의 작업 안에 1이 되는지를 `return`하는 함수 `coliatz`를 완성하세요

> 단! 500번을 넘어가도 1이 되지 않는다면, -1을 `return`할게요!

예시
```python
coliatz(6) #8
coliatz(16) #4
coliatz(626331) #-1
```

````python
def coliatz(n, times = 0):
    if times > 500:
        return -1
    elif n == 1:
        return times    
    else:
        if n % 2 == 0:
            return coliatz(n/2, times+1)
        else:
            return coliatz(n*3+1, times+1)
    
# def coliatz(num):
#     for i in range(500):
#         if num % 2:
#             num = num * 3 + 1
#         else:
#             num = num - 1
#         if num == 1:
#             return i + 1
#     return -1

coliatz(626331)
````
